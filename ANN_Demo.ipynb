```python
#Import packages
import numpy, random, os

#Initializations
lr = 1 #learning rate
bias = 1 #value of bias

#weights generated in a list (3 weights in total for 2 neurons and the bias)
weights = [random.random(),random.random(),random.random()] 
print('Before Training:',weights)

#Perceptron training-function (user-defined)
def Perceptron(input1, input2, output) :
    outputP = input1*weights[0]+input2*weights[1]+bias*weights[2]
    if outputP > 0 : #activation function (hardlim)
        pred = 1
    else :
        pred = 0
    error = output - pred
    weights[0] += error * input1 * lr
    weights[1] += error * input2 * lr
    weights[2] += error * bias * lr

#Training
iter = 50  #Training for 50 iterations
for i in range(iter) :
    Perceptron(1,1,1) #True or true
    Perceptron(1,0,1) #True or false
    Perceptron(0,1,1) #False or true
    Perceptron(0,0,0) #False or false

#Testing
x = float(input('First Input: '))
y = float(input('Second Input: '))
print('After Training: ',weights)

outputP = x*weights[0] + y*weights[1] + bias*weights[2]
print('Test Sum_k: ',outputP)

if outputP > 0 : #activation function (hardlim)
    pred = 1
else :
    pred = 0
print(x, "or", y, "is: ", pred)
```

```python
#Import packages
import numpy, random, os

#Initializations
lr = 1 #learning rate
bias = 1 #value of bias

#print('Before Training:',weights)

#Perceptron training-function (user-defined)
def Perceptron(input1, input2, output,weights) :
    outputP = input1*weights[0]+input2*weights[1]+bias*weights[2]
    if outputP > 0 : #activation function (hardlim)
        pred = 1
    else :
        pred = 0
    error = output - pred
    weights[0] += error * input1 * lr
    weights[1] += error * input2 * lr
    weights[2] += error * bias * lr
    
    return weights

#Training
epoch = 10  #Number of epoch
iter = 20000  #Training for 50 iterations
for j in range(epoch) :
    #weights generated in a list (3 weights in total for 2 neurons and the bias)
    w = [random.random(),random.random(),random.random()] 
    for i in range(iter) :
        w=Perceptron(1,1,1,w) #True or true
        #print('Weights at iter{',i+1,'}: ',weights)
        w=Perceptron(1,0,1,w) #True or false
        #print('Weights at iter{',i+1,'}: ',weights)
        w=Perceptron(0,1,1,w) #False or true
        #print('Weights at iter{',i+1,'}: ',weights)
        w=Perceptron(0,0,0,w) #False or false
        #print('Weights at iter{',i+1,'}: ',weights)
        #print('\n')

    #print('After Training:',weights)
    weights = w
    #Testing (fixed input)
    x = -2
    y = 2
    outputP = x*weights[0] + y*weights[1] + bias*weights[2]
    #print('Test Sum_k: ',outputP)

    if outputP > 0 : #activation function (hardlim)
        pred = 1
    else :
        pred = 0
    
    #Print formatting
    if j < 9 :
        print('Epoch: ',j+1,'  ',x, "or", y, "is : ", pred)
    else :
        print('Epoch: ',j+1,' ',x, "or", y, "is : ", pred)
```
